Mutation frequency per gene

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 50 genes by mutation frequency
top_genes = merged_df3['Gene'].value_counts().nlargest(50).index
filtered_df = merged_df3[merged_df3['Gene'].isin(top_genes)]

# Create the count plot
plt.figure(figsize=(18, 50))
sns.countplot(y='Gene', data=filtered_df, order=top_genes)
plt.title('Frequency of Mutations per Gene (Top 50 Genes)')
plt.xlabel('Count')
plt.ylabel('Gene')
plt.show()
```
